```python
print("\n--- Python OOP Project : Employee Management System ----")


class Person:
    def __init__(self):
        self.name = None
        self.age = None

    def set_person(self):
        self.name = input("Enter Name: ")
        self.age = int(input("Enter Age: "))

    def get_person(self):
        print(f"Person created: Name = {self.name}, Age = {self.age}")

    @staticmethod
    def display_all(person_list):
        print("\n---- All Person Details ----")
        for p in person_list:
            print(f"Name: {p.name}, Age: {p.age}")


class Employee:
    def __init__(self):
        self.name = None
        self.age = None
        self.employee_id = None
        self.salary = None
        self.department = None

    def set_employee(self):
        self.name = input("Enter Name: ")
        self.age = int(input("Enter Age: "))
        self.employee_id = int(input("Enter Employee ID: "))
        self.salary = float(input("Enter Salary: "))
        self.department = input("Enter Department: ")

    def get_employee(self):
        print(f"Employee created: Name={self.name}, Age={self.age}, "
              f"ID={self.employee_id}, Salary={self.salary}, Dept={self.department}")

    @staticmethod
    def display_all(emp_list):
        print("\n---- All Employee Details ----")
        for e in emp_list:
            print(f"Name: {e.name}, Age: {e.age}, ID: {e.employee_id}, Salary: {e.salary}, Dept: {e.department}")


class Manager:
    def __init__(self):
        self.name = None
        self.age = None
        self.employee_id = None
        self.salary = None
        self.department = None

    def set_manager(self):
        self.name = input("Enter Name: ")
        self.age = int(input("Enter Age: "))
        self.employee_id = int(input("Enter Employee ID: "))
        self.salary = float(input("Enter Salary: "))
        self.department = input("Enter Department: ")

    def get_manager(self):
        print(f"Manager created: Name={self.name}, Age={self.age}, "
              f"ID={self.employee_id}, Salary={self.salary}, Dept={self.department}")

    @staticmethod
    def display_all(manager_list):
        print("\n---- All Manager Details ----")
        for m in manager_list:
            print(f"Name: {m.name}, Age: {m.age}, ID: {m.employee_id}, Salary: {m.salary}, Dept: {m.department}")



Person_data = []
Employee_data = []
Manager_data = []


while True:
    print("\n--- Choose an Operation ---")
    print("1. Create Person")
    print("2. Create Employee")
    print("3. Create Manager")
    print("4. Show Details")
    print("5. Exit")

    choice = int(input("Enter your choice: "))

    match choice:

        case 1:
            num = int(input("Enter number of Persons: "))
            for _ in range(num):
                p = Person()
                p.set_person()
                p.get_person()
                Person_data.append(p)

        case 2:
            num = int(input("Enter number of Employees: "))
            for _ in range(num):
                e = Employee()
                e.set_employee()
                e.get_employee()
                Employee_data.append(e)

        case 3:
            num = int(input("Enter number of Managers: "))
            for _ in range(num):
                m = Manager()
                m.set_manager()
                m.get_manager()
                Manager_data.append(m)

        case 4:
            print("\n--- Choose Details to Show ---")
            print("1. Person")
            print("2. Employee")
            print("3. Manager")

            show_choice = int(input("Enter your choice: "))

            match show_choice:
                case 1:
                    Person.display_all(Person_data)
                case 2:
                    Employee.display_all(Employee_data)
                case 3:
                    Manager.display_all(Manager_data)

        case 5:
            print("Exiting system. Goodbye!")
            break

        case _:
            print("Invalid option! Try again.")

```
